```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0.5, 100.1, 0.5)


def func1(x):
    return x

def func2(x):
    return x**2

def func3(x):
    return (x**3) / 100

def func4(x):
    return np.sin(x)

def func5(x):
    return np.sin(x) / x

def func6(x):
  
    return np.log(x)

def func7(x):
    return np.exp(x)


fig, axs = plt.subplots(7, 1, figsize=(10, 20))  


axs[0].plot(x, func1(x), label='y = x')
axs[0].set_title('y = x')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(x, func2(x), label='y = x^2', color='orange')
axs[1].set_title('y = x^2')
axs[1].grid(True)
axs[1].legend()

axs[2].plot(x, func3(x), label='y = (x^3) / 100', color='green')
axs[2].set_title('y = (x^3) / 100')
axs[2].grid(True)
axs[2].legend()

axs[3].plot(x, func4(x), label='y = sin(x)', color='red')
axs[3].set_title('y = sin(x)')
axs[3].grid(True)
axs[3].legend()


x_non_zero = x[x != 0]
func5_values = func5(x_non_zero)

axs[4].plot(x_non_zero, func5(x_non_zero), label='y = sin(x) / x', color='purple')
axs[4].set_title('y = sin(x) / x')
axs[4].grid(True)
axs[4].legend()


x_non_zero_for_log = x[x > 0]
func6_values = func6(x_non_zero_for_log)

axs[5].plot(x_non_zero_for_log, func6(x_non_zero_for_log), label='y = log(x)', color='brown')
axs[5].set_title('y = log(x)')
axs[5].grid(True)
axs[5].legend()


subset_x = x[x <= 10]  
func7_values = func7(subset_x)

axs[6].plot(subset_x, func7(subset_x), label='y = e^x', color='blue')
axs[6].set_title('y = e^x')
axs[6].grid(True)
axs[6].legend()


plt.tight_layout()
plt.show()
```
